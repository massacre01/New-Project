```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
x
```

```python
y
```

```python
plt.plot(x,y,'r') #r is the color red
```

```python
plt.plot(x,y,'b') #r is the color red
plt.xlabel('Years Experience')
plt.ylabel('Income increase')
plt.title('Income and Experience')
plt.show()
```

```python
plt.plot(x,y,'s--') 
plt.xlabel('Years Experience')
plt.ylabel('Income increase')
plt.title('Income and Experience')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
#Create Figure (empty canvas)
fig = plt.figure()
#Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height

#Plot on the set of axes
axes.plot(x,y,'b')
axes.set_xlabel('Good People')
axes.set_ylabel('Bad People')
axes.set_title('Good v Bad')
```

```python
# Create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #inset axes

#Larger Figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('Years')
axes1.set_ylabel('Pollution')
axes1.set_title('Increase in pollution')

#Smaller Figure Axes 2 
axes2.plot(y,x,'r')
axes2.set_xlabel('Years')
axes2.set_ylabel('CO2 emissions')
axes2.set_title('CO2')
```

```python
#Use similar to plt.figure() except use tuple unpacking to grab fig and ax
fig, axes = plt.subplots()

#Now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Hello');
```
